```python
 import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
from scipy import stats
import numpy as np
```

```python
# Importing the Boston Housing dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df=pd.DataFrame(boston.data)
```

```python
df
```

```python
df
```

```python
df.columns = boston.feature_names
df.head()
```

```python
df.describe()
```

```python
#Adding target variable to dataframe
df['PRICE'] = boston.target 
df.head()
```

```python
# Spliting target variable and independent variables
X = df.drop(['PRICE'], axis = 1)
y = df['PRICE']
```

```python
# Splitting to training and testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 4)
```

```python
from sklearn.ensemble import RandomForestRegressor
train = RandomForestRegressor()
train.fit(X_train, y_train)
```

```python
# prediction train data
y_pred  = train.predict(X_train)
```

```python
# Model Evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

import sklearn.metrics as metrics
```

```python

print("Accuracy: ", train.score(X_test,y_test))
```

```python

print('R^2:',metrics.r2_score(y_train, y_pred))
print('MSE:',metrics.mean_squared_error(y_train, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
# Predicting Test data 
test_pred = train.predict(X_test)
```

```python
print('R^2:',metrics.r2_score(y_test, test_pred))
print('MSE:',metrics.mean_squared_error(y_test, test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```
